# **VALIDATION METHOD:-**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=34)
```

# FEATURE SCALING:-

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

# **ANN Model Building:-**

```python
#keras is a wrapper which is integrated with tensorflow.
from tensorflow.keras.models import Sequential # it provides the ann framework.
from tensorflow.keras.layers import Dense #helps to create different layer neurons.
from tensorflow.keras.layers import LeakyReLU,PReLU,ELU #activation function used in hidden and output layer.
from tensorflow.keras.layers import Dropout #used to deactivate nodes to avoid overfitting.
```

```python
classifier=Sequential() #installing the ann
```

```python
classifier.add(Dense(units=11,activation='relu')) #input layer
```

```python
classifier.add(Dense(units=6,activation='relu')) #1st hidden layer
```

```python
# classifier.add(Droupout(0.2)) #for drop- out
```

```python
classifier.add(Dense(units=6,activation='relu'))#2nd hidden layer
```

```python
classifier.add(Dense(units=1,activation='sigmoid')) #output layer
```

```python
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model_history=classifier.fit(x_train,y_train,validation_split=0.33,batch_size=10,epochs=50)
```

```python
y_pred=classifier.predict(x_test)
y_pred = (y_pred > 0.5)
```

```python
#confusion matrix
from sklearn.metrics import *
```

```python
confusion_matrix(y_test,y_pred)
```

```python
# Accuracy test:-
accuracy_score(y_test,y_pred)*100
```
